```python
class FavoriteSports:
    def __init__(self):
        self.sports = []

    def add_sport(self, sport):
        self.sports.append(sport)

    def view_sports(self):
        return self.sports

    def edit_sport(self, old_sport, new_sport):
        index = self.sports.index(old_sport)
        self.sports[index] = new_sport

    def delete_sport(self, sport):
        self.sports.remove(sport)

    def search_sport(self, sport):
        if sport in self.sports:
            return "Sport found in favorites"
        else:
            return "Sport not found in favorites"
```
